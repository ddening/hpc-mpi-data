### MPI_Barrier Runtime

```python
# =========================
#
# Titel : HPC Game Of Life
# Author: Dimitri Dening
# Date  : 29.06.2021
#
# =========================

import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
# reading the data from the file
file = "mpi_barrier_full.txt"

my_data = np.loadtxt(file)
x = my_data[:, 0]
y = my_data[:, 1]
```

```python
# plt.yscale('symlog')
plt.title("MPI_Barrier Runtime")
plt.xlabel("Threads")
plt.ylabel("Time in s")
plt.plot(y, x, "ro--")
plt.grid()
plt.tight_layout()
# plt.savefig('mpi-barrier-runtime.png')
plt.show()
```
